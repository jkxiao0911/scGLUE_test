```r
# 获取当前的列名
current_colnames <- colnames(my_data)
# # 使用正则表达式将列名中的前缀部分替换为空字符串
# new_colnames <- gsub("Mouse_Cerebellum_", "", current_colnames)

# # 更新 RangedSummarizedExperiment 对象的列名
# colnames(my_data) <- new_colnames
```

```r
# 获取数据矩阵
data_matrix <- assay(my_data)
# 将数据矩阵转换为稀疏矩阵
sparse_matrix <- as(data_matrix, "sparseMatrix")
```

```r
# 将稀疏矩阵的非零值设为1
binary_matrix <- as(sparse_matrix, "CsparseMatrix")  # 将稀疏矩阵转换为CsparseMatrix
binary_matrix@x[binary_matrix@x != 0] <- 1  # 将非零值设为1
```

```r
library(Seurat)
seurat_obj <- CreateSeuratObject(counts = binary_matrix)
obs_names <- colnames(seurat_obj)
```

```r
var_names <- rownames(seurat_obj)
```

```r
rna_assay <- seurat_obj[["RNA"]]

# 查看 "RNA" assay 的值（前100个行和列）
print(rna_assay[1:100, 1:100])
```

```r
#根据元数据对碱基序列进行细胞类型注释
meta <- read.delim("/mnt/disk1/xiaojk/data/mouse/Mouse_Cereb_ATAC_Cell_info.txt", header = TRUE)
obs_cell_types <- meta$Cell_type[match(obs_names, meta$Cell)]
cell_type_counts <- table(obs_cell_types)
sum(cell_type_counts)
seurat_obj$cell_type <- obs_cell_types
```

```r
#删除不必要的特征
seurat_obj$orig.ident <- NULL
seurat_obj$nCount_RNA <- NULL
seurat_obj$nFeature_RNA <- NULL
```

```r
# 提取列名包含 "sa001_Mouse_Cerebellum_e13_F" 或 "sa001_Mouse_Cerebellum_e13_M" 的数据
selected_cols <- colnames(seurat_obj)[grepl("sa001_Mouse_Cerebellum_e13_F|sa084_Mouse_Cerebellum_e13_M", colnames(seurat_obj))]

# 创建一个新的 Seurat 对象，仅包含选定的列
seurat_obj_e13 <- seurat_obj[, selected_cols]
```

```r
seurat_obj_e13
```

```r
library(SeuratDisk)
SaveH5Seurat(seurat_obj_e13, filename = "/mnt/disk1/xiaojk/data/mouse/peakmatrix_seurat_data_e13.h5Seurat")
Convert("/mnt/disk1/xiaojk/data/mouse/peakmatrix_seurat_data_e13.h5Seurat",dest = "h5ad")
```
