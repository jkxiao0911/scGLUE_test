```python
section_id ='151673'    
input_dir = os.path.join('/mnt/disk1/xzhou/data/spatial_3D/DLPFC/', section_id)
rna = sc.read_visium(path=input_dir, count_file=section_id + '_filtered_feature_bc_matrix.h5', load_images=True)
rna.var_names_make_unique(join="++")

# read the annotation
Ann_df = pd.read_csv(os.path.join(input_dir, section_id + '_truth.txt'), sep='\t', header=None, index_col=0)
Ann_df.columns = ['Ground Truth']
Ann_df[Ann_df.isna()] = "unknown"
rna.obs['Ground Truth'] = Ann_df.loc[rna.obs_names, 'Ground Truth'].astype('category') 

#'Ground Truth' 是原始列名，你需要将其改为 'cell_type'
rna.obs.rename(columns={'Ground Truth': 'cell_type'}, inplace=True)
```

```python
chromStart_columns = ["chromStart", "chromStart"]
chromEnd_columns = ["chromEnd", "chromEnd"]
chrom_columns = ["chrom", "chrom"]
strand_columns = ["strand", "strand"]
# 创建新的整合列
rna.var["chromStart1"] = rna.var[chromStart_columns].apply(lambda row: next((val for val in row if pd.notna(val)), None), axis=1)
rna.var["chromEnd1"] = rna.var[chromEnd_columns].apply(lambda row: next((val for val in row if pd.notna(val)), None), axis=1)
rna.var["chrom1"] = rna.var[chrom_columns].apply(lambda row: next((val for val in row if pd.notna(val)), None), axis=1)
rna.var["strand1"] = rna.var[strand_columns].apply(lambda row: next((val for val in row if pd.notna(val)), None), axis=1)
# 删除原先的属性列
rna.var.drop(columns=chromStart_columns + chromEnd_columns + chrom_columns + strand_columns, inplace=True)
rna.var.rename(columns={"chromStart1": "chromStart", "chromEnd1": "chromEnd", "chrom1": "chrom", "strand1": "strand"}, inplace=True)
rna.var.loc[:, ["chrom", "chromStart", "chromEnd","strand"]]
```

```python
var_df = pd.DataFrame(rna.var)

# 检查是否存在缺失值的行
missing_rows = var_df[var_df[["chromStart", "chromEnd"]].isnull().any(axis=1)]

# 从 rna 中删除缺失的 var 变量
rna = rna[:, ~rna.var.index.isin(missing_rows.index)]

valid_chroms = ['chr' + str(i) for i in range(1, 23)] + ['chrX', 'chrY']

# 检查 chrom 列是否包含有效的字符串值
mask = rna.var['chrom'].isin(valid_chroms)

# 从 rna 中删除 chrom 列不包含有效字符串值的 var 变量
rna = rna[:, mask]
```

```python
# # 要删除的特征名称列表
# features_to_drop = ["feature_types", "genome", "gene_ids","hgnc_id","havana_gene","artif_dupl"]

# # 删除多个特征
# rna.var.drop(columns=features_to_drop, inplace=True)

rna.var = rna.var.loc[:, ~rna.var.columns.duplicated(keep='first')]
```

```python
# 假设您的 DataFrame 名称为 rna，列名为 "strand"
not_nan_mask = rna.var["strand"].notna()

# 使用布尔掩码来选择非 NaN 数据的行
non_nan_strand_data = rna.var[not_nan_mask]

# 打印非 NaN 数据的行
print(non_nan_strand_data["strand"])
```

```python
from itertools import chain
guidance_hvf = guidance.subgraph(chain(
    rna.var.query("highly_variable").index,
    atac.var.query("highly_variable").index
)).copy()
```

```python
glue = scglue.models.fit_SCGLUE(
    {"rna": rna, "atac": atac}, guidance_hvf,
    fit_kws={"directory": "glue"}
)
```

```python
glue.save("glue.dill")
# glue = scglue.models.load_model("glue.dill")
```

```python
dx = scglue.models.integration_consistency(
    glue, {"rna": rna, "atac": atac}, guidance_hvf
)
dx
```

```python
import seaborn as sns
_ = sns.lineplot(x="n_meta", y="consistency", data=dx).axhline(y=0.05, c="darkred", ls="--")
```

```python
import numpy as np
rna.obsm["tangram_ct_pred"] = pd.DataFrame(neigh.predict_proba(rna.obsm["X_glue"]),
                                           index=rna.obs_names, columns=np.sort(list(atac.obs["cell_type"].unique())))
```

```python
def construct_obs_plot(df_plot, adata, perc=0, suffix=None):
    # clip
    df_plot = df_plot.clip(df_plot.quantile(perc), df_plot.quantile(1 - perc), axis=1)

    # normalize
    df_plot = (df_plot - df_plot.min()) / (df_plot.max() - df_plot.min())

    if suffix:
        df_plot = df_plot.add_suffix(" ({})".format(suffix))
    adata.obs = pd.concat([adata.obs, df_plot], axis=1)
    
    
def plot_cell_annotation_sc(
    adata_sp, 
    annotation_list, 
    x="x", 
    y="y", 
    spot_size=None, 
    scale_factor=None, 
    perc=0,
    alpha_img=1.0,
    bw=False,
    ax=None
):
        
    # remove previous df_plot in obs
    adata_sp.obs.drop(annotation_list, inplace=True, errors="ignore", axis=1)

    # construct df_plot
    df = adata_sp.obsm["tangram_ct_pred"][annotation_list]
    construct_obs_plot(df, adata_sp, perc=perc)
    
    #non visium data 
    if 'spatial' not in adata_sp.obsm.keys():
        #add spatial coordinates to obsm of spatial data 
        coords = [[x,y] for x,y in zip(adata_sp.obs[x].values,adata_sp.obs[y].values)]
        adata_sp.obsm['spatial'] = np.array(coords)
    
    if 'spatial' not in adata_sp.uns.keys() and spot_size == None and scale_factor == None:
        raise ValueError("Spot Size and Scale Factor cannot be None when ad_sp.uns['spatial'] does not exist")
    
    #REVIEW
    if 'spatial' in adata_sp.uns.keys() and spot_size != None and scale_factor != None:
        raise ValueError("Spot Size and Scale Factor should be None when ad_sp.uns['spatial'] exists")
    
    sc.pl.spatial(
        adata_sp, color=annotation_list, cmap="viridis", show=False, frameon=False, spot_size=spot_size,
        scale_factor=scale_factor, alpha_img=alpha_img, bw=bw, ax=ax, colorbar_loc=None,ncols=5,wspace=0
    )

    adata_sp.obs.drop(annotation_list, inplace=True, errors="ignore", axis=1)
```
